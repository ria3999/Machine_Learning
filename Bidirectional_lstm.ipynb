```
#@title Default title text
def remove_pattern(input_txt, pattern):
    r = re.findall(pattern, input_txt)
    for i in r:
        input_txt = re.sub(i, '', input_txt)        
    return input_txt  


df['tweet'] = np.vectorize(remove_pattern)(df['tweet'], "@[\w]*")
df['tweet'] = df['tweet'].str.replace("[^a-zA-Z#]", " ")
```

```
import tensorflow as tf
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
max_features=20000
embed_dim = 100
```

```
tokenizer = Tokenizer(num_words=max_features,oov_token='<oov_tok>')
tokenizer.fit_on_texts(df['tweet'])
word_index = tokenizer.word_index
print(word_index)
```

```
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dropout,Bidirectional
from tensorflow.keras.layers import Embedding
from tensorflow.keras.layers import Conv1D
from sklearn.feature_extraction.text import CountVectorizer
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split
from keras.utils.np_utils import to_categorical
import re
```

```
import gensim.downloader as api
glove_model=api.load('glove-wiki-gigaword-300')
```

```
embedding_index={}
with open("/content/glove.twitter.27B.100d.txt") as f:
    for line in f:
        values=line.split()
        word=values[0]
        coefs=np.asarray(values[1:])
        embedding_index[word]=coefs
embedding_matrix=np.zeros((max_features+1,embed_dim))

for word,i in word_index.items():
    embedding_vector=embedding_index.get(word)
    if embedding_vector is not None:
        embedding_matrix[i]=embedding_vector
print(len(embedding_matrix))
```

```
embedding_matrix.shape
```

```
max_features
```

```
#model.add(Bidirectional(LSTM(20, return_sequences=True), input_shape=(n_timesteps, 1)))

#lstm_out = 200
model = Sequential()
model.add(Embedding(max_features+1, embed_dim,weights=[embedding_matrix],input_length = padded.shape[1]))
model.add(Bidirectional(LSTM(16,return_sequences=True)))
#model.add(Conv1D(128,5,activation="relu"))
model.add(Bidirectional(LSTM(32)))
model.add(Dense(24,activation="relu"))
#model.add(Bidirectional(LSTM(lstm_out,dropout=0.5)))
model.add(Dense(2,activation='softmax'))
model.compile(loss = 'binary_crossentropy', optimizer='adam',metrics = ['accuracy'])
print(model.summary())
```

```
model.fit(X1_train, Y1_train, epochs = 5,validation_data=[X1_test,Y1_test])
```

```
test_loss,test_acc=model.evaluate(X1_test,Y1_test)
```
